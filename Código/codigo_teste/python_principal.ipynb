```python
 df_padrao = pd.read_csv("DADOS_PROJETO/Data/Data/Escolas/escolas122023.csv", sep=";", encoding='latin1')
```

```python
columns_padrao = df_padrao.columns
diretorio_escola = "DADOS_PROJETO/Data/Data/Escolas"
arquivo_csv_escolas = os.listdir(diretorio_escola)
```

```python
lista_dataframe = []
for arquivo in arquivo_csv_escolas:
    if '.csv' in arquivo:
        caminho_completo = diretorio_escola + '/' + arquivo
        df = pd.read_csv(caminho_completo, names= columns_padrao, header= None, sep=";", encoding='latin1', skiprows=1) 
        lista_dataframe.append(df)
    else:
        continue       
```
